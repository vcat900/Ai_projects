```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
import tensorflow as tf
import pandas as pd
```

```python
model = tf.keras.applications.VGG16(
    include_top=True,
    weights="imagenet",
    input_tensor=None,
    input_shape=None,
    pooling=None,
    classes=1000,
    classifier_activation="softmax",
)
```

```python
model.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])
```

```python
model.save('Vgg.h5')
```
