```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
import tensorflow as tf
import pandas as pd
```

```python
(x_train, y_train), (x_test, y_test) = tf.keras.datasets.fashion_mnist.load_data()
```

```python
# Reshape input data to include a single channel
x_train = x_train.reshape(x_train.shape[0], 28, 28, 1)
x_test = x_test.reshape(x_test.shape[0], 28, 28, 1)
```

```python
x_train = x_train / 255
x_test = x_test / 255
```

```python
labels = ['T-shirt/top', 'Trouser', 'Pullover', 'Dress', 'Coat', 'Sandal', 'Shirt', 'Sneaker', 'Bag', 'Ankle boot']
```

```python
model = tf.keras.Sequential()
model.add(tf.keras.layers.Conv2D(32, 1, input_shape=x_train.shape[1:]))
model.add(tf.keras.layers.MaxPooling2D())
model.add(tf.keras.layers.Flatten())
model.add(tf.keras.layers.Dense(392, activation='relu'))
model.add(tf.keras.layers.Dense(196, activation='relu'))
model.add(tf.keras.layers.Dense(10))
```

```python
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
history = model.fit(x_train, y_train, epochs=14)
```

```python

model.save('Fashion.h5')
```
